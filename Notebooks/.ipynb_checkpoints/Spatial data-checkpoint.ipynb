# Choose the area based on the fisrt and last code

This can be Gebieden, Buurten or Wijken

```python

area = 'wijken'

if area == 'buurten':
    begin = 'A00a'
    end = 'Z99z'
elif area == 'wijken':
    begin = 'A00'
    end = 'Z99'
elif area == 'gebieden':
    begin = 'DX01'
    end = 'ZX99'

x = 0

wijken = []

for i in df['gebiedcode15'].unique():

    if i == begin:
        x = 1
    
    if i == end:
        break
        
    if x == 1:
        wijken.append(i)

print(len(wijken))
```

# Get amount of persons per area


```python
df_t = df[df['variabele'] == 'BEVDICHT']

df_t = df_t[df_t['jaar'] == 2019]
df_t = df_t.loc[df['gebiedcode15'].isin(wijken)]

df_t['amount_of_persons'] = (df_t['waarde']/df_t.waarde.sum()) * 861000

df_t
```

# Get area distribution based on ethnicity

```python
groepen = ['BEVSUR_P',
'BEVANTIL_P',
'BEVTURK_P',
'BEVMAROK_P',
'BEVOVNW_P',
'BEVWEST_P',
'BEVAUTOCH_P',
]

df_sp = df[df['jaar'] == 2019]


df_sp = df_sp.loc[df['variabele'].isin(groepen)]

df_sp = df_sp.loc[df['gebiedcode15'].isin(wijken)]

df_sp = df_sp.replace(['BEVANTIL_P',
'BEVOVNW_P',
'BEVWEST_P'], 'OVERIG')

df_sp = df_sp.groupby(['variabele', 'gebiedcode15'], as_index=False).sum()

df_sp = df_sp.replace(['BEVAUTOCH_P','BEVSUR_P','BEVMAROK_P','BEVTURK_P', 'OVERIG' ],['Autochtoon','Surinaams','Marokkaans','Turks', 'Overig'])


amount = []

for i in df_sp.iterrows():
    row = i[1]
    area = row['gebiedcode15']
    waarde = row['waarde']
    etn = row['variabele']
    
    print(i)
    total = df_t[df_t['gebiedcode15'] == area]['amount_of_persons']
    print(total)
    if math.isnan(total):
        amount.append(0)
        continue
        
    if int(total) == 0:
        amount.append(0)
        continue
    

    amount.append(math.ceil((float(total)*float(waarde)/100)) )

df_sp['amount_per_area'] = amount


df_sp['area_code'] = [i%len(wijken) for i in range(len(wijken)*5)]
df_sp = df_sp.fillna(0)
df_sp

# df_sp.to_csv('../Data/Spatial_data/Ethnic_data.csv')

```

# Get Area distribution based on gender

```python
groepen = ['BEVMAN_P', 'BEVVROUW_P']
df_sp = df[df['jaar'] == 2019]


df_sp = df_sp.loc[df['variabele'].isin(groepen)]

df_sp = df_sp.loc[df['gebiedcode15'].isin(wijken)]
df_sp = df_sp.replace(['BEVMAN_P','BEVVROUW_P'],['Man','Vrouw'])

df_sp = df_sp.groupby(['variabele', 'gebiedcode15'], as_index=False).sum()


amount = []

for i in df_sp.iterrows():
    row = i[1]
    area = row['gebiedcode15']
    waarde = row['waarde']
    etn = row['variabele']
    
    total = df_t[df_t['gebiedcode15'] == area]['amount_of_persons']
    
    print(total)
    if math.isnan(total):
        amount.append(0)
        continue
        
    if int(total) == 0:
        amount.append(0)
        continue
    amount.append(math.ceil((float(total)*float(waarde)/100)) )

df_sp['amount_per_area'] = amount

print(sum(amount))

df_sp['area_code'] = [i%len(wijken) for i in range(len(wijken)*2)]



df_sp.to_csv('../Data/Spatial_data/Gender_data.csv')
```

# Make one dataframe based on all three distributions

```python
gender_data = pd.read_csv('../Data/Spatial_data/Gender_data.csv')

etn_data = pd.read_csv('../Data/Spatial_data/Ethnic_data.csv')

edu_data = pd.read_csv('../Data/Spatial_data/Edu_data_withbias.csv')
codes = []
etnicities = []
amounts = []
percentages = []
genders = []
educations = []

df = pd.DataFrame()
for i in etn_data.itertuples():
#     print(i)
    e = i[2]
    c = i[-1]

    amount = i[-2]
    
    waarde = round(i[-3]/100,100)
    
    
    for j in ['Man', 'Vrouw']:
        
        gender = gender_data[gender_data['area_code'] == c ]
        gender = gender[gender['variabele'] == j]
        gender_waarde = round(float((gender['waarde'])/100),100)

        for k in [1,2,3]:
            
            education = edu_data[edu_data['area_code'] == c]
            education = education[education['variabele'] == k]
            
           

            print(amount, (education['amount_per_area']),)
            amount2 = amount + float(education['amount_per_area']) + float(gender['amount_per_area'])
        
            amounts.append(amount2)
            codes.append(c)
#             percentages.append(waarde2)
            etnicities.append(e)
            genders.append(j)
            educations.append(k)
        
df['etngrp'] = etnicities
df['code'] = codes
df['geslacht'] = genders
df['amount'] = amounts
df['oplniv'] = educations
df['percentage'] = percentages
```
